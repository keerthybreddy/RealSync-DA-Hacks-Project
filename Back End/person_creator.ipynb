Now, let's put all of thesse into an equation

```python
def person_creator(tuple_data):
    import pandas as pd
    import os
    class Person(object):
    # Constructor
        def __init__(self, name='', grad_year=18, major='',quarter='',
                    pronoun = '',hobbies='',fun_fact='',food=''
                    ,clubs='', res=''):
            self.name = name
            self.grad_year = grad_year
            self.major = major
            self.quarter = quarter
            self.pronoun = pronoun
            self.hobbies = hobbies
            self.fun_fact = fun_fact
            self.food = food
            self.clubs = clubs
            self.res = res
    test_person = Person(*tuple_data)
    user_pd = pd.DataFrame(columns=['Key','Value'])
    for attr, value in test_person.__dict__.items():
            key = attr
            value = value
            user_pd = user_pd.append({'Key':key ,'Value':value}, ignore_index = True)
    splitted = test_person.name.split(" ")
    join = ""
    for item in splitted:
        join += item
    return user_pd.to_json('.\processed_people\%s.json' %join)
```

```python
person_creator(data)
person_creator(data1)
person_creator(data2)
person_creator(data3)
person_creator(data4)
person_creator(data5)
person_creator(data6)
```
